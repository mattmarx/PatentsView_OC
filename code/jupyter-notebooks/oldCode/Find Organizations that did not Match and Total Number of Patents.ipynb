# Match Patents with Assignee_id

```python
### start timer
t0=time.time()

### set the path for the input file and save to variable
res_folder = "../sourceFiles/"
input_file = "patent_assignee.tsv"
a_full=os.path.join(res_folder,input_file)
print(a_full,"\n")

### imports the dfMergedFullDataSet file to merge onto Mike's output file using the ID column
patAssignee=pd.read_csv(a_full,sep="\t")

### end timer and print total time
t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

### print general stats and first 5 records for dataset
display(patAssignee.info(null_counts=True),patAssignee.head())
```

```python
mergePatents=patAssignee.merge(noMatchStateless2,on=['assignee_id'],how='inner')

display(mergePatents.info(),mergePatents.head())
```

```python
counts=pd.DataFrame(mergePatents.groupby(by=['assignee_id'],
                                         as_index=False)['patent_id'].count()).sort_values(by=['patent_id'],ascending=False)
counts1=counts.loc[counts['patent_id'] > 1].reset_index(drop=True)

display(counts1.info(),counts1.head())
```

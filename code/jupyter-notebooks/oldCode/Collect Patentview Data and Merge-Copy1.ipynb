# Make dataset - no reassignment

```python
import pandas as pd
import time
import warnings
warnings.filterwarnings('ignore')

t0=time.time()

# loads assignee dataset and filters for US company or corp (2) and foreign company or corp (3)
assignee=pd.read_csv("../sourceFiles/assignee.tsv",sep='\t')
assignee=assignee.loc[(assignee['type']==2) | (assignee['type']==3)]

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(assignee.info(null_counts=True),assignee.head())
```

```python
t0=time.time()

# loads location dataset for businesses and individuals
location=pd.read_csv("../sourceFiles/location.tsv",sep='\t')

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(location.info(null_counts=True),location.head())
```

```python
t0=time.time()

# loads patent_assignee dataset and is used as a xwalk between the assignee and location datasets
patAssignee=pd.read_csv("../sourceFiles/patent_assignee.tsv",sep='\t')

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

print(len(patAssignee['patent_id'].unique()))
display(patAssignee.info(null_counts=True),patAssignee.head())
```

```python
t0=time.time()

# loads patent dataset to include the date the patent was granted
patDate=pd.read_csv("../sourceFiles/patent.tsv",sep='\t').iloc[:,[2,4,10]]

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(patDate.info(null_counts=True),patDate.head())
```

```python
# 0 = not withdrawn and 1 = withdrawn
patDate['withdrawn'].value_counts()
```

```python
t0=time.time()

# add the patent date to the assignee/location data
patAssigneeDate=patAssignee.merge(patDate,left_on='patent_id',right_on='number',how='inner')
patAssigneeDate=patAssigneeDate.iloc[:,[0,1,2,4]]

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(patAssigneeDate.info(null_counts=True),patAssigneeDate.head())
```

```python
t0=time.time()

# find the total number of patents per location
patPerLoc=patAssigneeDate.groupby(['location_id','assignee_id'],as_index=False)['patent_id'].count()
patPerLoc=pd.DataFrame(patPerLoc)
patPerLoc.rename(columns={"patent_id": "patCnt"},inplace=True)

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(patPerLoc.info(),patPerLoc.head())
```

```python
t0=time.time()

# find the earliest patent date for assignee
patEarDate=patAssigneeDate.sort_values(['patent_id','assignee_id'],ascending=False).groupby(['assignee_id'],as_index=False)['date'].min()
patEarDate=pd.DataFrame(patEarDate)
# patEarDate.rename(columns={"patent_id": "patCnt"},inplace=True)

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(patEarDate.info(),patEarDate.head())
```

```python
t0=time.time()

# remove the patent feature and drop assignee and location ids
patAssignee1=patAssignee.iloc[:,1:]
patAssignee2=patAssignee1.drop_duplicates(subset=['assignee_id','location_id'])

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(patAssignee2.info(null_counts=True),patAssignee2.head())
```

```python
t0=time.time()

# print the number of nulls and non-nulls for the location_id feature
print(len(patAssignee2[patAssignee2['location_id'].notnull()]))
print(len(patAssignee2[patAssignee2['location_id'].isnull()]))

# select non-null records for the location_id feature
patAssigneeNoNulls=patAssignee2[patAssignee2['location_id'].notnull()]

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(patAssigneeNoNulls.info(null_counts=True),patAssigneeNoNulls.head())
```

```python
t0=time.time()

# merge the patent counts per location with patent_assignee dataset
patCntsPerLoc=patAssigneeNoNulls.merge(patPerLoc,on=['location_id','assignee_id'],how='inner')

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(patCntsPerLoc.info(null_counts=True),patCntsPerLoc.head())
```

```python
t0=time.time()

# merge the granted patent dates with patent_assignee dataset
patCntsPerLocDates=patCntsPerLoc.merge(patEarDate,on=['assignee_id'],how='inner')

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(patCntsPerLocDates.info(null_counts=True),patCntsPerLocDates.head())
```

```python
t0=time.time()

# performs merge between assignee and patent_assignee datasets; this places a location_id
# feature in the dataset that can be utilized to include the location from the locations data
df1=assignee.merge(patCntsPerLocDates,left_on='id',right_on='assignee_id',how='inner')

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(df1.info(null_counts=True),df1.head())
```

```python
t0=time.time()

# merge the location data with the previous dataset
df2=df1.merge(location,left_on='location_id',right_on='id',how='inner')

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(df2.info(null_counts=True),df2.head())
```

```python
t0=time.time()

# select the columns
df=df2.iloc[:,[2,3,5,6,7,1,4,10,11,12,15,8]]

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(df.info(null_counts=True),df.head())
```

```python
t0=time.time()

# find the number of non-nulls and nulls in the name_first feature
print(len(df[df['name_first'].notnull()]))
print(len(df[df['name_first'].isnull()]))

# make a new dataset with the name_first feature having null values; this removes individuals
# from the resulting dataset
dfNulls=df[df['name_first'].isnull()]

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(dfNulls.info(null_counts=True),dfNulls.head())
```

```python
t0=time.time()

# reduce the number of columns that are not necessary
dfNullsNoInd=dfNulls.iloc[:,2:]

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(dfNullsNoInd.info(null_counts=True),dfNullsNoInd.head())
```

```python
t0=time.time()

# removes null values from organization to allow .str.contains function below to filter out
# universities and medical
dfNullsNoIndNa=dfNullsNoInd.dropna(subset=['organization'])

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(dfNullsNoIndNa.info(null_counts=True),dfNullsNoIndNa.head())
```

```python
t0=time.time()

# removes universities and hospitals; list came from Matt
dfNullsNoIndNaUniMed=dfNullsNoIndNa[dfNullsNoIndNa['organization'].str.contains('Barts|Berkeley|Coll|Dept|Ecole|Hopkins|Hsch|\
                                 Max Planck|Med Sch|Politec|Polytec|School|Suny|Ucl|Umc|Univ|Institute|university|University|\
                                 faculty|Mit|Caltech|Nyu|Ucl|Univ|College|Ucsd|Cmu|Ucbl|Epfl|Lmu|Interuniv|Harvard|Case|\
                                 Western|Caltech|Cern|Chalmers|Cuny|Mit|Nyu|Riken|Th Darmstadt|Uc Davis|Ucsf|Usc|\
                                 Virginia Tech|Georgia Tech|Trustees Of',case=False)==False]

dfNullsNoIndNaUniMed=dfNullsNoIndNaUniMed.reset_index(drop=True)

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(dfNullsNoIndNaUniMed.info(null_counts=True),dfNullsNoIndNaUniMed.head())
```

```python
t0=time.time()

# total the number of patents per organization by grouping the assignee_id
sumPatsPerAssignee=dfNullsNoIndNaUniMed.groupby(['assignee_id'],as_index=False)['patCnt'].sum()
sumPatsPerAssignee.rename(columns={"patCnt": "sum"},inplace=True)

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(sumPatsPerAssignee.info(null_counts=True),sumPatsPerAssignee.head())
```

```python
t0=time.time()

# add the total number of patents to the dfNullsNoIndNaUniMed dataset
dfNullsNoIndNaUniMed1=dfNullsNoIndNaUniMed.merge(sumPatsPerAssignee,on='assignee_id',how='inner')

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(dfNullsNoIndNaUniMed1.info(null_counts=True),dfNullsNoIndNaUniMed1.head())
```

```python
t0=time.time()

# calculate the percentage each assignee contributes to the overall patents for the organization
dfNullsNoIndNaUniMed1['percent']=(dfNullsNoIndNaUniMed1['patCnt']/dfNullsNoIndNaUniMed1['sum'])*100
dfNullsNoIndNaUniMed1['percent']=dfNullsNoIndNaUniMed1['percent'].round(1)

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(dfNullsNoIndNaUniMed1.info(null_counts=True),dfNullsNoIndNaUniMed1.head())
```

```python
dfNullsNoIndNaUniMed1.rename(columns={"sum": "patCntSumPerAssignee","percent": "percentPatCntSumPerAssignee",
                                      "date": "dateOfFirstPat"},inplace=True)
```

# Add Number of Locations per Assignee

```python
import pandas as pd
import time
import warnings
warnings.filterwarnings('ignore')

t0=time.time()

# loads assignLoc dataset
df=pd.read_csv("../csvResults/assignLoc.csv").iloc[:,1:]

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(df.info(null_counts=True),df.head())
```

```python
t0=time.time()

# counts the number of locations per assignee
df1=df.groupby(['assignee_id'],as_index=False)['location_id'].count()
df1.rename(columns={'location_id':'totalLocations'},inplace=True)

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(df1.info(null_counts=True),df1.head())
```

```python
t0=time.time()

# appends the assigneeLoc dataset with the total number of locations per assignee
df2=df.merge(df1,on='assignee_id',how='inner')

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(df2.info(null_counts=True),df2.head())
```

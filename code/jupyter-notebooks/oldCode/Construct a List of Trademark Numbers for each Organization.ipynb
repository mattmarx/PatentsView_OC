# Construct a Dataset Containing Trademark Numbers for All Assignees

```python
### import the libraries used to process the PatentsView and OC data.
import pandas as pd
import time
import os
import warnings
warnings.filterwarnings('ignore')

### start timer
t0=time.time()

### set the path for the input file and save to variable
source_folder = "../sourceFiles/"
input_file = "tm_assignee.csv"
a_full=os.path.join(source_folder,input_file)

assigneeName=pd.read_csv(a_full).iloc[:,[0,1,4,5,8]]
assigneeName.rename(columns={"ee_name":"organization","ee_city":"city","ee_state":"state",
                             "ee_natlty":"ctrlEntity"},inplace=True)

### end timer and print total time
t1 = time.time()
total = t1-t0
print("Total time is %4f" % (total/60), "mins")

### print general stats and first 5 records for dataset 
display(assigneeName.info(null_counts=True),assigneeName.head())
```

```python
### start timer
t0=time.time()

### set the path for the input file and save to variable
source_folder = "../sourceFiles/"
input_file = "tm_cf_no.csv"
a_full=os.path.join(source_folder,input_file)

serialNo=pd.read_csv(a_full).iloc[:,[0,1]]

### end timer and print total time
t1 = time.time()
total = t1-t0
print("Total time is %4f" % (total/60), "mins")

### print general stats and first 5 records for dataset 
display(serialNo.info(null_counts=True),serialNo.head())
```

```python
### start timer
t0=time.time()

assigneeSerialNo=assigneeName.merge(serialNo,on=['rf_id'],how='inner')

### end timer and print total time
t1 = time.time()
total = t1-t0
print("Total time is %4f" % (total/60), "mins")

### print general stats and first 5 records for dataset 
display(assigneeSerialNo.info(),assigneeSerialNo.head())
```

```python
### start timer
t0=time.time()

assigneeSerialNo1=assigneeSerialNo.iloc[:,[0,5,1,2,3,4]].reset_index(drop=True)

### end timer and print total time
t1 = time.time()
total = t1-t0
print("Total time is %4f" % (total/60), "mins")

### print general stats and first 5 records for dataset 
display(assigneeSerialNo1.info(),assigneeSerialNo1.head())
```

```python
### start timer
t0=time.time()

# assigneeSerialNo2=assigneeSerialNo1.groupby(by=['rf_id','organization'])['serial'].unique()
assigneeSerialNo2=assigneeSerialNo1.groupby(by=['organization'])['serial'].unique()
assigneeSerialNo3=pd.DataFrame(assigneeSerialNo2).reset_index()

### end timer and print total time
t1 = time.time()
total = t1-t0
print("Total time is %4f" % (total/60), "mins")

### print general stats and first 5 records for dataset 
display(assigneeSerialNo3.info(),assigneeSerialNo3.head())
```

```python
### start timer
t0=time.time()

assigneeSerialNo4=assigneeSerialNo1.iloc[:,[0,2,3,4,5]]
assigneeSerialNo5=assigneeSerialNo4.drop_duplicates(keep='first').reset_index(drop=True)
assigneeSerialNo5.dropna(subset=['city','state'],inplace=True)

### end timer and print total time
t1 = time.time()
total = t1-t0
print("Total time is %4f" % (total/60), "mins")

### print general stats and first 5 records for dataset 
display(assigneeSerialNo5.info(show_counts=True),assigneeSerialNo5.head())
```

```python
### start timer
t0=time.time()

assgnWithSerial=assigneeSerialNo3.merge(assigneeSerialNo5,on=['organization'],how='inner')
assgnWithSerial['organization']=assgnWithSerial['organization'].str.title()
assgnWithSerial['city']=assgnWithSerial['city'].str.title()
assgnWithSerial['state']=assgnWithSerial['state'].str.title()
assgnWithSerial['ctrlEntity']=assgnWithSerial['ctrlEntity'].str.title()

### end timer and print total time
t1 = time.time()
total = t1-t0
print("Total time is %4f" % (total/60), "mins")

### print general stats and first 5 records for dataset 
display(assgnWithSerial.info(),assgnWithSerial.head())
```

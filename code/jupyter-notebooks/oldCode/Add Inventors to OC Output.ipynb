# Make dataset - no reassignment

```python
import pandas as pd
import time
import warnings
warnings.filterwarnings('ignore')

t0=time.time()

# load the inventor data
inv=pd.read_csv("../sourceFiles/inventor.tsv",sep='\t').iloc[:,0:3].reset_index(drop=True)

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(inv.info(null_counts=True),inv.head())
```

```python
import pandas as pd
import time
import warnings
warnings.filterwarnings('ignore')

t0=time.time()

# load the xwalk file that includes inventor location and patent ids
patInv=pd.read_csv("../sourceFiles/patent_inventor.tsv",sep='\t')

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(patInv.info(null_counts=True),patInv.head())
```

```python
t0=time.time()

# loads location dataset for individuals
location=pd.read_csv("../sourceFiles/location.tsv",sep='\t').iloc[:,0:4]
location=location.loc[location['country']=='US'].reset_index(drop=True)
location=location.rename(columns={'id':'location_id'})

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(location.info(null_counts=True),location.head())
```

```python
t0=time.time()

# loads patent dataset to include the date the patent was granted
patDate=pd.read_csv("../sourceFiles/patent.tsv",sep='\t').iloc[:,[2,4,10]]

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(patDate.info(null_counts=True),patDate.head())
```

```python
# 0 = not withdrawn and 1 = withdrawn
patDate['withdrawn'].value_counts()
```

```python
t0=time.time()

invLoc=inv.merge(patInv,left_on=['id'],right_on=['inventor_id'],how='outer',indicator=True)
invLoc1=invLoc.loc[invLoc['_merge']=='both'].iloc[:,[0,1,2,3,5]].reset_index(drop=True)
invLoc1.dropna(subset=['name_first','name_last','location_id'],inplace=True)

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(invLoc1.info(null_counts=True),invLoc1.head())
```

```python
t0=time.time()

invLoc1=invLoc1.rename(columns={'id':'inventor_id'})

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(invLoc1.info(null_counts=True),invLoc1.head())
```

```python
t0=time.time()

invLocState=invLoc1.merge(location,on=['location_id'],how='outer',indicator=True)
invLocState=invLocState.loc[invLoc['_merge']=='both'].iloc[:,0:8]
invLocState.dropna(subset=['inventor_id','name_first','name_last','patent_id','location_id','state','country'],inplace=True)
invLocState=invLocState.reset_index(drop=True)

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(invLocState.info(null_counts=True),invLocState.head())
```

```python
t0=time.time()

invLocStateDate=invLocState.merge(patDate,left_on=['patent_id'],right_on=['number'],how='outer',indicator=True)
invLocStateDate=invLocStateDate.loc[invLoc['_merge']=='both'].iloc[:,[0,1,2,3,4,5,6,7,9]]
invLocStateDate.dropna(subset=['inventor_id','name_first','name_last','patent_id','location_id','state','country',
                               'date'],inplace=True)
invLocStateDate=invLocStateDate.reset_index(drop=True)

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(invLocStateDate.info(null_counts=True),invLocStateDate.head())
```

```python
t0=time.time()

invLocStateDate1=invLocStateDate.drop_duplicates(subset=['inventor_id','patent_id','location_id'])
invLocStateDate1=invLocStateDate1.reset_index(drop=True)

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(invLocStateDate1.info(null_counts=True),invLocStateDate1.head())
```

```python
t0=time.time()

invLocStateMinDate=invLocStateDate1.sort_values(['patent_id','inventor_id'],
                                                 ascending=False).groupby(['inventor_id'],as_index=False)['date'].min()
invLocStateMinDate=invLocStateMinDate.drop_duplicates(subset=['inventor_id'])
invLocStateMinDate=invLocStateMinDate.reset_index(drop=True)

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(invLocStateMinDate.info(null_counts=True),invLocStateMinDate.head())
```

```python
t0=time.time()

dfMinDate=invLocStateDate1.merge(invLocStateMinDate,on=['inventor_id','date'],how='inner')

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(dfMinDate.info(null_counts=True),dfMinDate.head())
```

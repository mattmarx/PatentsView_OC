```python
import pandas as pd
import time
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

t0=time.time()

df=pd.read_csv("../sourceFiles/location.tsv",sep="\t").iloc[:,0:6]
df.rename(columns={"id":"location_id"},inplace=True)

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(df.info(null_counts=True),df.head())
```

```python
df_sub=df.loc[df['country']=='US'].reset_index(drop=True)
df_sub['city']=df_sub['city'].str.lower()
df_sub['state']=df_sub['state'].str.lower()

display(df_sub.info(null_counts=True),df_sub.head())
```

```python
t0=time.time()

dfMatt=pd.read_csv("../sourceFiles/zipcitystatelatlon.csv").iloc[:,[1,2,3,4]]

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(dfMatt.info(null_counts=True),dfMatt.head())
```

```python
dfMatt1=dfMatt.drop_duplicates(subset=['city','state'],keep='first').reset_index(drop=True)
dfMatt1
```

```python
dfMatt1['state'].unique()
```

```python
dfMatt2=dfMatt1[dfMatt1['state'].str.contains("PR|VI|AS|nan") == False].reset_index(drop=True)
dfMatt2
```

```python
dfMatt2['city']=dfMatt2['city'].str.lower()
dfMatt2['state']=dfMatt2['state'].str.lower()

dfMatt2
```

```python
df_sub1=df_sub[df_sub['latitude'].isnull()]
df_sub1
```

```python
dfMerge=df_sub1.merge(dfMatt2,on=['city','state'],how='inner',suffixes=('_matt','_loc'))
dfMerge
```

```python
dfMerge1=dfMerge.iloc[:,[0,1,2,3,6,7]]
dfMerge1.rename(columns={'latitude_loc':"latitude","longitude_loc":"longitude"},inplace=True)
dfMerge1
```

```python
df_sub2=df_sub[df_sub['latitude'].notnull()]
df_sub2
```

```python
supData=pd.concat([df_sub2,dfMerge1],axis=0)
supData
```

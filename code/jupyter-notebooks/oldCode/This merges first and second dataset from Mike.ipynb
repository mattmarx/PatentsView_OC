```python
### remove the location_id from the coordinate set obtained from patentsview because the
### address_city and agent_city from OC does not have a location_id to merge the data.
### NAs present in the city and state were also removed because it cause incorrect pairing
### during the subsequent merging steps

latLong1=latLong.iloc[:,1:]
latLong1.dropna(subset=['city','state'],inplace=True)

display(latLong1.info(),latLong1.head())
```

```python
### merge the lat/long coordinates data with the patentsview/OC data using city and states

latLong2=latLong1.merge(cityLatLongPatView,left_on=['city','state'],right_on=['address_city','address_state'],
                        how='outer',suffixes=('_LatLong','_patView'),indicator=True)
latLong2.rename(columns={'latitude_LatLong':'address_city_latitude','longitude_LatLong':'address_city_longitude',
                         'city_patView':'city','state_patView':'state'},
                inplace=True)
latLong2.drop(labels=['city_LatLong','state_LatLong'],axis=1,inplace=True)

display(latLong2.info(),latLong2.head())
```

```python
### reordering data

eighteen_col = latLong2.pop('address_city_latitude')
nineteen_col = latLong2.pop('address_city_longitude')

latLong2.insert(16, 'address_city_latitude', eighteen_col)
latLong2.insert(17, 'address_city_longitude', nineteen_col)

display(latLong2.info(),latLong2.head())
```

```python
latLong2Left=latLong2.loc[latLong2['_merge']=='right_only']

display(latLong2Left.info(),latLong2Left.head())
```

```python
### concatenate the data where records are found in both sets and when the patentsview/OC data
### do not have a matching city and/or state; this retains those records so they will not be lost

latLong2Concat=pd.concat([latLong2Both,latLong2Left],axis=0).iloc[:,0:22].sort_values(by=['ID']).reset_index(drop=True)

display(latLong2Concat.info(),latLong2Concat.head())
```

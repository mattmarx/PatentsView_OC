```python
import pandas as pd
import time
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

t0=time.time()

# 
df=pd.read_csv("../csvResults/assignLocTot.csv").iloc[:,1:]

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(df.info(null_counts=True),df.head())
```

```python
t0=time.time()

df1=df.loc[df['country']=='US'].iloc[:,[0,1,2,3,4,5,6,9,10,11,12]].reset_index(drop=True)

t1=time.time()
total=t1-t0
print("Total time is %4f" % (total/60), "mins")

display(df1.info(null_counts=True),df1.head())
```

```python
# filter records that are 100% of patent holders
df100=df1.loc[df1['percentPatCntSumPerAssignee']==100]

display(df100.info(),df100.head())
```

```python
# construct set that is less than 100%
dfNot100=df1.loc[df1['percentPatCntSumPerAssignee']!=100]

display(dfNot100.info(),dfNot100.head())
```

```python
# set of if statements to group assignees if the total number of patents are 100 or less
dfNot100.loc[dfNot100['patCntSumPerAssignee'] < 101, '100 or less'] = 1
dfNot100.loc[dfNot100['patCntSumPerAssignee'] >=101, '100 or less'] = 0

dfNot100
```

```python
# all records should be kept for a label of 1 under the 100 or less column 
dfNot100_1=dfNot100.loc[dfNot100['100 or less']==1].iloc[:,0:11]

# records need further processing for a label of 0 under the 100 or less column 
dfNot100_0=dfNot100.loc[dfNot100['100 or less']==0].iloc[:,0:11]
```

```python
display(dfNot100_1.info(),dfNot100_1.head())
```

```python
dfNot100_0=dfNot100_0.sort_values(by=['assignee_id','percentPatCntSumPerAssignee'],ascending=False)

display(dfNot100_0.info(),dfNot100_0.head())
```

```python
dfNot100_0_8099=dfNot100_0.loc[dfNot100_0['percentPatCntSumPerAssignee']>=80]

display(dfNot100_0_8099.info(),dfNot100_0_8099.head())
```

```python
dfNot100_0_492=dfNot100_0.loc[(dfNot100_0['percentPatCntSumPerAssignee']<80) & \
                              (dfNot100_0['percentPatCntSumPerAssignee']>=50) & (dfNot100_0['totalLocations']>2)]

display(dfNot100_0_492.info(),dfNot100_0_492.head())
```

```python
theRest1=dfNot100_0.merge(dfNot100_0_8099,how='outer',indicator=True)
theRest2=theRest1.loc[theRest1['_merge']=='left_only'].iloc[:,0:11]

display(theRest2.info(),theRest2.head())
```

```python
theRest3=theRest2.merge(dfNot100_0_492,how='outer',indicator=True)
theRest4=theRest3.loc[theRest3['_merge']=='left_only'].iloc[:,0:11]

display(theRest4.info(),theRest4.head())
```
